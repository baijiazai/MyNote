```python
import jieba
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
from PIL import Image
from imageio import imread
```

```python
# 全模式
seg_list = jieba.cut('我来到北京清华大学', cut_all=True)
print('，'.join(seg_list))
```

```python
# 精准模式
seg_list = jieba.cut('我来到北京清华大学', cut_all=False)
print('，'.join(seg_list))
```

```python
# 搜索引擎模式
seg_list = jieba.cut_for_search('小明硕士毕业于中国科学院计算所，后在日本京都大学深造')
print('，'.join(seg_list))
```

```python
content = open(r'C:\Users\Administrator\python\dataset\pachong.txt', encoding='utf-8').read()
Counter(content).most_common(10)
```

```python
con_words = [x for x in jieba.cut(content) if len(x) >= 2]
Counter(con_words).most_common(10)
```

```python
# 词云图
new_text = ' '.join(jieba.cut(content))
wordcloud = WordCloud(font_path='simhei.ttf', background_color='black').generate(new_text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
new_text = ' '.join(jieba.cut(content))
pac_mask = imread(r'C:\Users\Administrator\python\dataset\apchong.png')
wc = WordCloud(font_path='simhei.ttf', background_color='white', max_words=2000, mask=pac_mask).generate(new_text)
plt.imshow(wc)
plt.axis('off')
plt.show()
# wc.to_file(file_name) 保存词云图
```
